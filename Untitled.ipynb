```python
import requests

def fetchAndSaveTextToFile(url, path, encoding='utf-8'):
    """Fetches text content from a URL and saves it to a file.

    Args:
        url (str): The URL of the content to fetch.
        path (str): The path to the file where the content will be saved.
        encoding (str, optional): The encoding of the content. Defaults to 'utf-8'.

    Raises:
        requests.exceptions.RequestException: If the request encounters an error.
    """

    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception for non-200 status codes

        # Open the file in text mode ("w") with the specified encoding
        with open(path, "w", encoding=encoding) as f:
            f.write(response.text)
        print(f"Text content saved to: {path}")

    except requests.exceptions.RequestException as e:
        print(f"Error downloading content: {e}")

url = "https://timesofindia.indiatimes.com/city"
fetchAndSaveTextToFile(url, "data/times.html")
```
